```python
import random

# Define the size of the grid
grid_size = 5

# Initialize player and treasure positions
player_position = [random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)]
treasure_position = [random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)]

# Maximum number of moves
max_moves = 10
moves_made = 0

# Function to calculate Manhattan distance
def calculate_distance(pos1, pos2):
    return abs(pos1[0] - pos2[0]) + abs(pos1[1] - pos2[1])

# Initial distance
initial_distance = calculate_distance(player_position, treasure_position)

print(f"Initial player position: {player_position}")
print(f"Treasure position: {treasure_position}")
print(f"You have {max_moves} moves to find the treasure.")

while moves_made < max_moves:
    move = input("Enter your move (N, S, E, W): ").strip().upper()

    # Record old position
    old_position = player_position.copy()
    
    # Move player
    if move == 'N':
        if player_position[1] > 0:
            player_position[1] -= 1
        else:
            print("Move not allowed. Out of bounds!")
            continue
    elif move == 'S':
        if player_position[1] < grid_size - 1:
            player_position[1] += 1
        else:
            print("Move not allowed. Out of bounds!")
            continue
    elif move == 'E':
        if player_position[0] < grid_size - 1:
            player_position[0] += 1
        else:
            print("Move not allowed. Out of bounds!")
            continue
    elif move == 'W':
        if player_position[0] > 0:
            player_position[0] -= 1
        else:
            print("Move not allowed. Out of bounds!")
            continue
    else:
        print("Invalid move! Use N, S, E, or W.")
        continue

    # Increment move counter
    moves_made += 1
    
    # Calculate new distance
    new_distance = calculate_distance(player_position, treasure_position)

    # Check if player found the treasure
    if player_position == treasure_position:
        print(f"Congratulations! You found the treasure at {player_position} in {moves_made} moves!")
        break
    else:
        if new_distance < initial_distance:
            print("You're getting closer to the treasure!")
        else:
            print("You're moving farther from the treasure!")

        print(f"Current position: {player_position}")
        print(f"Moves left: {max_moves - moves_made}")

if player_position != treasure_position:
    print(f"Game over! You ran out of moves. The treasure was at {treasure_position}.")
```
